## **Actual EvapoTranspiration**
The GEOET component offers the possibility to compute the **Actual Evapotranspiration** considering different potential evapotranspiration model. 
The Actual Evapotranspiration is computed multiplying the Potential Evapotranspiration for a total stress factor $g_{s}$. The stress factor is computed using the model proposed by Macfarlane et al. (2004), but instead of using the stress factor given by the leaf water potential, we used the one based on the soil moisture proposed by FAOapproach (Allen, 1986).

\begin{equation}
 ET_{eff}=ET_{pot}\cdot g_{s}
\end{equation}

\begin{equation}
 g_{s} = g_{s}max\cdot f(R_{par})\cdot f(T_{a})\cdot f(VPD) \cdot K_{s}
\end{equation}

where:
- $g_{s}max $ is the conductance without any kind of stress and in well-watered conditions;
- $f(R_{par})$ is the stress coefficient as a function of photosynthetically active radiation;
- $f(T_{a})$ is the stress coefficient as a function of the air temperature;
- $f(VPD) $ is the stress coefficient as a function of the vapor pressure deficit;
- $K_{s} $ is the stress coefficient as a function of the water content.

---
### **Water Stress**

The **water stress coefficient** $K_{s}$ is computed using the [(FAO formulation)](https://www.fao.org/4/X0490E/x0490e0e.htm#water%20stress%20coefficient%20(ks)):


\begin{equation}
 K_{s} = \frac{TAW - D_{r}}{TAW-RAW}
\end{equation}

where

- $TAW$ is the **total available soil water** in the root zone [mm];
- $D_{r}$ is the **root zone depletion**, water shortage relative to field capacity, thus if $\theta = \theta_{FC}$, $D_r =0$ ;
- $RAW$ is the **readily available soil water** in the root zone [mm],the fraction of TAW that a crop can extract from the root zone without suffering water stress.


The **total available soil water** $TAW$ is computed according the following equation: 

\begin{equation}
 TAW = 1000 (\theta_{FC}-\theta_{WP})\cdot Z_{r}
\end{equation}

where

- $\theta_{FC}$ is the water content at **field capacity** [-] [(Table 19)](https://www.fao.org/4/X0490E/x0490e0c.htm#TopOfPage:~:text=in%20Table%2019.-,TABLE%2019,-.%20Typical%20soil%20water);
- $\theta_{WP}$ is the water content at **wilting point** [-] [(Table 19)](https://www.fao.org/4/X0490E/x0490e0c.htm#TopOfPage:~:text=in%20Table%2019.-,TABLE%2019,-.%20Typical%20soil%20water);
- $Z_{r}$ is the **root depth** [(Table 22)](https://www.fao.org/4/X0490E/x0490e0e.htm#water%20stress%20coefficient%20(ks):~:text=in%20Figure%2041.-,TABLE%2022.%20Ranges,-of%20maximum%20effective).

The **readily available soil water** $RAW$ is computed according the following equation: 

\begin{equation}
RAW = p \cdot TAW 
\end{equation}

where
- $p$ is the average fraction of Total Available Soil Water (TAW) that can be depleted from the root zone before moisture stress (reduction in ET) occurs [0-1].

[Here (Table 22)](https://www.fao.org/4/X0490E/x0490e0e.htm#water%20stress%20coefficient%20(ks):~:text=in%20Figure%2041.-,TABLE%2022.%20Ranges,-of%20maximum%20effective), you can find value of the **depletation fraction** for different crops.

### **Air Temperature Stress**

The **air temperature stress factor** can be computed as:


\begin{equation}
 f_{T} = b(T_{a}-T_{low})(T_{up}-T_{a})^c
\end{equation}

where b and c are defined as:


\begin{equation}
 c = \frac{T_{up} - T_{opt}}{T_{opt} - T_{low}}
\end{equation}

\begin{equation}
 b= \frac{1}{(T_{opt} - T_{low})(T_{up} - T_{opt})^c}
\end{equation}

- $T_{opt}$ is the **temperature** at maximum conductance [C];
- $T_{low}$ and $T_{up}$ the **lower and upper temperature** of the range for which a positive stomatal conductance is predicted[C].

If \begin{equation}
 T_{low} < T_{air} < T_{up} \\\ f(T) = 0 
\end{equation}

White et al. (1999) assigned the values for $T_{low}$ , $T_{opt}$ and $T_{up}$ equal to 0C, 17C and 38C. These parameters can be set a priori or calibrated.

### **Solar Radiation Stress**

The **solar radiation stress** can be computed as:


\begin{equation}
 f(R_{sw}) = (\frac{1}{2\theta}(\alpha\ R_{sw}+1-\sqrt{ (\alpha\ R_{sw}+1)^2 -4\theta \alpha\ R_{sw}  }))^{-1}
\end{equation}

where $\alpha$ and $\theta$ are the slope and shape parameters of the stress function $f(R_{par})$ and are set equal to 0.005 and 0.85 [-].



--- 
### **Vapour Pressur Deficit Stress**

The **vapour pressure deficit stress** can be computed as:


\begin{equation}
 f(VPD) = 1.1\cdot exp(-0.63\cdot VPD)
\end{equation}
